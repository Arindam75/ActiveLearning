```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
sns.set(style="darkgrid")
```

### Learning Curves for Full Training Set 

```python
df_full = pd.read_csv('./logger/trainlog_baseline.csv')
df_full.head()
```

```python
fig1, ax1 = plt.subplots(1,2,figsize=(18, 4))

sns.lineplot(x=df_full['epoch'], y=df_full['categorical_accuracy'], data=df_full, ax = ax1[0])
sns.lineplot(x=df_full['epoch'], y=df_full['val_categorical_accuracy'], data=df_full, ax = ax1[0])
ax1[0].set_title('Accuracy')

sns.lineplot(x=df_full['epoch'], y=df_full['loss'], data=df_full, ax = ax1[1])
sns.lineplot(x=df_full['epoch'], y=df_full['val_loss'], data=df_full, ax = ax1[1])
ax1[1].set_title('Loss')
```

```python
df_al = pd.read_csv('./logger/trainlog_al.csv')
df_al.head()
```
